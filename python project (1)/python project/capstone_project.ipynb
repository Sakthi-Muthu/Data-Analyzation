# MACHINE LEARNING TECHNIQUES

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(max_depth=3)
```

```python
dt.fit(x_train,y_train)
```

```python
print(dt.score(x_train,y_train))
print(dt.score(x_test,y_test))
```

```python
y_pred=dt.predict(x_test)
```

```python
print(y_pred)
```

```python
from sklearn import metrics
```

```python
r_square=metrics.r2_score(y_test,y_pred)
print('R-square error for decision tree regreessor is:',round(100*(r_square),3))
```

```python
y_pred=dt.predict(x_test)
from sklearn import metrics
r_square=metrics.r2_score(y_test,y_pred)
print('R=square error associated eith decision tree regressor is:',round(100*(r_square),2))
```

# RANDOM FOREST REGRESSOR

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
rfr=RandomForestRegressor(random_state=5)
rfr.fit(x_train,y_train)
```

```python
y1_pred=rfr.predict(x_test)
```

```python
print('Mean Absolute Error (MAE):',metrics.mean_absolute_error(y_test,y1_pred))
print('Mean Squared Error (MSE):,',metrics.mean_squared_error(y_test,y1_pred))
print('Root Mean Squared Error (RMSE):',
      np.sqrt(metrics.mean_squared_error(y_test,y1_pred)))
mape=np.mean(np.abs((y_test-y1_pred)/np.abs(y1_pred)))
mape1=round(mape*100,2)
accuracy1=round(100*(1-mape),2)
print('Mean Absolute Percentage Error (MAPE):',round(mape*100,2))
print('Accuracy:',round(100*(1-mape),2))
```
