# Prepare the data frame for further analysis

```python
# Step 1 check for null values, drop it and cross check 
df.isnull()
```

```python
df1=df.dropna()
```

```python
df1.isnull().sum()
```

```python
# Step 2 check for duplicate values and drop it
df1.duplicated()
```

```python
df1.drop_duplicates()
```

```python
# Step 3 rectify column name
df1.rename(columns={'Price Each':'price of each items'},inplace=True)
df1.head()
```

```python
# Step 4 Extract date from order date
#df1['Order Date'] = pd.to_datetime(df1['Order Date'])
#df1['extracted date'] = df1['Order Date'].dt.date
#print(df1[['Order Date','extracted date']])
#df1['date']=df1['Order Date'].str[:8]
#df1
```

```python
# Step 4 Extract date from order date
df1['Extracted Date'] = df1['Order Date'].str[:8]
df1
```

```python
# Step 5 extract month from order date

#df1['Order Date'] = pd.to_datetime(df1['Order Date'])


#df1['order_month'] = df1['Order Date'].dt.month


#print(df1[['Order Date', 'order_month']])
```

```python
# Step 5 extract month from order date
df1['month']=df1['Order Date'].str[:2]
df1
```

```python
#Step 6 extract house_no ,city and pin code from address
df1[['House No','City','Pin code']]=df1['Purchase Address'].str.split(',',expand=True)
df1
```

```python
# Step 7 Check for month column for bad data entry and remove it
df1.month.value_counts()

```

```python
df1=df1[df1['month'].str[:2] != 'Or']
df1.head()
```

```python
# Step 8 correct data types
df1.dtypes
```

```python
# Step 8 correct data types
df1['month']=df1['month'].astype('int32')
df1['Order Date']=pd.to_datetime(df1['Order Date'])
df1['Extracted Date']=pd.to_datetime(df1['Extracted Date'])
df1.info()
```

## Step 9 make Sales Column (hint Quantity_ordered * Price_each)



```python
df1['Quantity Ordered']=df1['Quantity Ordered'].astype("int32")
df1['price of each items']=df1['price of each items'].astype("float")
df1["Sales"]=df1['Quantity Ordered'] * df1['price of each items']
df1
```

# Sales Value by Each Month

```python
df1.groupby('month')['Sales'].sum().reset_index()
```

# Question 5 Find out the distribution of sales over different times of the day (Morning,Afternoon, Evening, Night) and answer the following for each city.

```python
df1['Hour']=df1['Order Date'].dt.hour
```

```python
df1.groupby('Hour').count()
```

```python
hours=[hour for hour,df1 in df1.groupby('Hour')]
plt.plot(hours,df1.groupby('Hour').count())
plt.grid(True)
plt.xticks(hours)
plt.xlabel('Hours in 24 Hr format')
plt.ylabel('Order')
plt.show()
```

# Question 6 Highest Product Sold by Amount and Numbers

```python
Highest_order=df1.Product.value_counts()
Highest_sales=df1.groupby('Product').Sales.sum().sort_values(ascending=False)
fig,axes=plt.subplots(1,2,figsize=(20,7))
sns.barplot(ax=axes[0],x=Highest_order.index,y=Highest_order.values)
axes[0].set_xlabel('Products')
axes[0].set_ylabel('No. of Orders')
axes[0].tick_params(axis='x',rotation=90)
sns.barplot(ax=axes[1],x=Highest_sales.index,y=Highest_sales.values)
axes[1].set_xlabel('Products')
axes[1].set_ylabel('Sales Amount')
axes[1].tick_params(axis='x',rotation=90)
plt.show()
```
