```python
# rename column iso 3166-1 alpha 3 CODE to Iso_Code
new_column_name = 'Iso_Code'
df.rename(columns={' iso 3166-1 alpha 3 CODE': new_column_name}, inplace=True)
df


```

```python
# top 5 countries which have more number of covid cases (hint use df.nlargest)
 top_countries= df.nlargest(5, 'Total Cases')

# Display the top 5 countries
print("\nTop 5 Countries by COVID-19 Cases:")
print(top_countries)
```

```python
#plot bar graph for top 5 countries
plt.figure(figsize=(10, 6))
plt.bar(top_countries['Country'], top_countries['Total Cases'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Number of COVID-19 Cases')
plt.title('Top 5 Countries with Highest COVID-19 Cases')
plt.show()
```

```python
# least 5 countries which have least number of covid cases
Least_countries= df.nsmallest(5, 'Total Cases')

# Display the top 5 countries
print("\nLeast 5 Countries by COVID-19 Cases:")
print(Least_countries)
```

```python
#draw bar graph of least 5 countries
plt.figure(figsize=(10, 6))
plt.bar(Least_countries['Country'], top_countries['Total Cases'], color='lavender')
plt.xlabel('Country')
plt.ylabel('Number of COVID-19 Cases')
plt.title('Least 5 Countries with COVID-19 Cases')
plt.show()

```

```python
#top 5 countries which have more number of covid Deaths
topD_countries= df.nlargest(5, 'Total Deaths')

# Display the top 5 countries
print("\nTop 5 Countries by COVID-19 deaths Cases:")
print(topD_countries)
```

```python
#plot bar graph for top 5 countries which have more number of covid Deaths
plt.figure(figsize=(10, 6))
plt.bar(topD_countries['Country'], topD_countries['Total Deaths'], color='red')
plt.xlabel('Country')
plt.ylabel('Number of  Cases')
plt.title('Top 5 Countries with covid deaths rate')
plt.show()
```

```python
#5 countries which have least number of covid Death
least_deaths_countries= df.nsmallest(5, 'Total Deaths')

# Display the top 5 countries
print("\n 5 Countries by COVID-19 deaths Cases:")
print(least_deaths_countries)
```

```python
#Death percentage
#top 5 countries which have more number of covid  Death percentage 
topP_countries= df.nlargest(5, 'Death percentage')

# Display the top 5 countries
print("\nTop 5 Countries by COVID-19 Cases:")
print(topP_countries)
```

```python
#plot bar graph for top 5 countries which have more number of covid  Death percentage 
plt.figure(figsize=(10, 6))
plt.bar(topP_countries['Country'], topP_countries['Death percentage'], color='red')
plt.xlabel('Country')
plt.ylabel('Death percentage')
plt.title('top 5 countries which have more number of covid  Death percentage ')
plt.show()
```

```python
#continent wise cases (hint use group by)
continent_cases=df.groupby('Continent')['Total Cases'].sum().reset_index()
print(continent_cases)
```

```python
#draw bar plot for continent wise case
plt.figure(figsize=(10, 6))
plt.bar(continent_cases['Continent'],continent_cases['Total Cases'], color='red')
plt.xlabel('Continent')
plt.ylabel('Total Cases')
plt.title('continent wise case ')
plt.show()
```
